```python
import requests
URL="https://quotes.toscrape.com/"

res=requests.get(URL)
```

```python
print(res)
```

```python
with open("scraped_data/data3.html" ,"w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
from bs4 import BeautifulSoup
with open("scraped_data/data3.html","r",encoding="utf-8") as f:
    html_content1=f.read()
    
soup=BeautifulSoup(html_content1,"lxml")
```

```python
quotes=soup.select("div.quote")
```

```python
print(quotes)
```
