```python

import os
os.environ["OMP_NUM_THREADS"] = "4"
import seaborn as sns
from sklearn.datasets import make_blobs

```

```python
X,y=make_blobs(
  n_samples=1000,
  n_features=2,
  centers=2,
   random_state=42
)
```

```python
X
```

```python
#k-means
from sklearn.cluster import KMeans
```

```python
K=4
kmean=KMeans(
    n_clusters=K,
    random_state=42
)
```

```python
labels=kmean.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],c=labels)
```

```python
#choosing k value

#1elbow method;;;;
wcss=[]
for k in range(1,21):
    kmea=KMeans(n_clusters=K)
    kmea.fit_predict(X)
    wcss.append(kmea.inertia_)
```

```python
sns.lineplot(x=range(1,21),y=wcss,marker='o')
```
